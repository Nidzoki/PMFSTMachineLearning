Napišite funkciju **Multiply** koja kao ulazne parametre prima prirodne brojeve $a$, $b$, $c$ i $d$  te kao rezultat vraća ***umnožak matrica*** dimenzija $a\times b$  te  $c\times d$ (ako je umnožak moguće izračunati).   Svaka matrica je reprezentirana jednim numpy arrayem kojem su elementi slučajno generirani  cijeli brojevi iz raspona $[1,7 \rangle$. Ako umnožak nije moguće izračunati, potrebno je ispisati poruku o tome.

> Napomena: Negdje ispišite generirane matrice.



```
import numpy as np
import random as rn

def Multiply(a, b, c, d):
  return "Nemoguće izračunati" if b != c else np.dot(np.array([[rn.randint(1,7) for i in range(b)] for j in range(a)]), np.array([[rn.randint(1,7) for i in range(d)] for j in range(c)]))
```

```
C = Multiply(2, 3, 3, 5)

print(f"C: {C}")
print(f"C.shape = {C.shape}")
```

```
C = Multiply(2, 3, 4, 5)
print(C)
print(C.shape)
```

## Zadatak 1.
 Napišite funkciju **RandomArray** koja kao ulazne parametre prima dva prirodna broja $m, n$ i string $d \in \{ \text{"uniform", "normal", "poisson"} \}$ te kao rezultat vraća $m\times n$ numpy array slučajnih brojeva generiranih iz distribucije $d$.

<br>

> *Napomena:*
>
>
>Parametre distribucije
> - $\mu$ i $\sigma$ kod normalne
> - granice $min$ i $max$ kod uniformne
> - $\lambda$ kod Poissonove
>
> unosi korisnik (unos unutar metode).

Ako nije poslan "ispravan" parametar $d$, onda se brojevi generiraju iz standardne normalne distribucije ($\mu = 0$, $\sigma = 1$).












Kao pomoć pri rješavanju zadataka koristite [NumPy dokumentaciju](https://docs.scipy.org/doc/numpy-1.16.0/reference/index.html).

```
import numpy as np

def RandomArray(m, n, d):

  if d == "uniform":
    min_val = float(input("Unesi minimalnu vrijednost za uniformnu distribuciju: "))
    max_val = float(input("Unesi maksimalnu vrijednost za uniformnu distribuciju: "))
    return np.random.uniform(min_val, max_val, size=(m, n))
  elif d == "normal":
    mu = float(input("Unesi mean (mu) za normalnu distribuciju: "))
    sigma = float(input("Unesi standardnu devijaciju (sigma) za normalnu distribuciju: "))
    return np.random.normal(mu, sigma, size=(m, n))
  elif d == "poisson":
    lam = float(input("Unesi lambdau za Poissonovu distribuciju: "))
    return np.random.poisson(lam, size=(m, n))
  else:
    print("Pogrešan parametar. Koristi se standardna normalna distribucija (mu=0, sigma=1).")
    return np.random.normal(0, 1, size=(m, n))
```

### Test

```
m, n = 5, 7
```

```
A = RandomArray(m, n, "nn")
print(A)
print(A.shape)
```

```
A = RandomArray(m, n, "poisson")
print(A)
print(A.shape)
```

```
A = RandomArray(m, n, "normal")
print(A)
print(A.shape)
```

```
A = RandomArray(m, n, "uniform")
print(A)
print(A.shape)
```

## Zadatak 2

Pomoću funkcije iz *Zadatka 1* generirajte dva $5 \times 100000$ numpy arrayja **A** i **B** iz normalne distribucije s parametrima
- **A**: $\mu = 50$ i $\sigma = 2$
- **B**: $\mu = -50$ i $\sigma = 2$.

Izračunajte srednje vrijednost **po retcima** za **A**, **B** i **A + B**.
Što uočavate?

Uočavam da su distribucije skoro pa jednake jer smo imali veliku matricu.

```
# A: mu = 50 i sigma = 2
A = RandomArray(5, 100000, "normal")
# B: mu = -50 i sigma = 2
B = RandomArray(5, 100000, "normal")
```

```
print(f"srednja vrijednost po retcima A:", np.mean(A, axis=1))
print(f"srednja vrijednost po retcima B:", np.mean(B, axis=1))
print(f"srednja vrijednost po retcima A + B:", np.mean(A + B, axis=1))
```
