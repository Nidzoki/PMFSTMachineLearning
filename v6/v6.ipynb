```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
data.info()
```

```python
nez_atr = data.columns[:-1]
zav_atr = data.columns[-1]
X = data[nez_atr]
y = data[zav_atr]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

### Zadatak 1.1

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.tree import plot_tree
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
plot_tree(tree)
```

### Zadatak 1.2

```python
print("Točnost stabla odluke na skupu za treniranje: ", round(metrics.accuracy_score(y_train, tree.predict(X_train)), 2) * 100, "%")
print("Točnost stabla odluke na skupu za testiranje: ", round(metrics.accuracy_score(y_test, tree.predict(X_test)), 2) * 100, "%")
```

Model je prenaučen

### Zadatak 1.3.1

```python
tree_depth_3 = DecisionTreeClassifier()
tree_depth_3.max_depth = 3
tree_depth_3.fit(X_train, y_train)
plot_tree(tree_depth_3)
```

```python
print("Točnost na skupu za treniranje: ", round(metrics.accuracy_score(y_train, tree_depth_3.predict(X_train)), 2) * 100, "%")
print("Točnost na skupu za testiranje: ", round(metrics.accuracy_score(y_test, tree_depth_3.predict(X_test)), 2) * 100, "%")
```

Nepodrezano stablo je prenaučeno, a podrezano nije dovoljno naučeno 😆

### Zadatak 1.4

```python
from sklearn.model_selection import ParameterGrid
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'max_depth': range(3, 20), 'min_samples_leaf': range(2, 30)}
```

```python
search = GridSearchCV(tree_depth_3, parameters)
```

```python
search.fit(X_train, y_train)
```

```python
print("Najbolji parametri: ", search.best_params_)
```

```python
grid_search_tree = search.best_estimator_
grid_search_tree.fit(X_train, y_train)
```

```python
print("Točnost na skupu za testiranje: ", round(metrics.accuracy_score(y_test, grid_search_tree.predict(X_test)), 2)*100, "%")
```

### Zadatak 1.5

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier()
random_forest.fit(X_train, y_train)
```

```python
print("Točnost na skupu za testiranje: ", round(metrics.accuracy_score(y_test, random_forest.predict(X_test)), 2)*100, "%")
```

```python
X_part = data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Outcome']]
X_part.info()
```

```python
Y_part = data['Age']
```

```python
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_part, Y_part, test_size=0.3, random_state=2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train_2, y_train_2)
```

### Zadatak 2.2

```python
plot_tree(regressor)
```

### Zadatak 2.3

```python
print("MAE: ", round(metrics.mean_absolute_error(y_test_2, regressor.predict(X_test_2)), 2))
print("MSE: ", round(metrics.mean_squared_error(y_test_2, regressor.predict(X_test_2)), 2))
print("RMSE: ", round(metrics.root_mean_squared_error(y_test_2, regressor.predict(X_test_2)), 2))
print("R2: ", round(metrics.r2_score(y_test_2, regressor.predict(X_test_2)), 2))
```
